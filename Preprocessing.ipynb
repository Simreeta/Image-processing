Preprocessing of Image


Data set used is from skimage

Gray scale image

```
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2gray

original = data.camera()
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

Gaussian filter:
It is used for removing noise from the image

```
from scipy import misc,ndimage
   
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

camera = data.camera()  
camera1 = data.camera() 
blurred_face = ndimage.gaussian_filter(camera1, sigma=1)  
very_blurred = ndimage.gaussian_filter(camera1, sigma=0)
#Results  
ax[0].imshow(camera)
ax[0].set_title("Original")
ax[1].imshow(camera1)
ax[1].set_title("Gaussian applied")

fig.tight_layout()
plt.show()
```

Mask the object

```
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
   
image = data.camera()  
type(image)

mask = image < 87  
image[mask]=255  

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(camera)
ax[0].set_title("Original")
ax[1].imshow(image, cmap='gray')
ax[1].set_title("Object detection")

fig.tight_layout()
plt.show()
```

Sobel edge detection in gray scale

```
import matplotlib.pyplot as plt  
%matplotlib inline
       
from skimage import data,filters
       
image = data.camera() 
edges = filters.sobel(image)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(camera)
ax[0].set_title("Original")
ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Sobel edge detection")

fig.tight_layout()
plt.show()
```

Perwitt edge detection

```
import matplotlib.pyplot as plt  
%matplotlib inline
       
from skimage import data,filters
       
camera = data.camera()
edges = filters.prewitt(camera)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(camera)
ax[0].set_title("Original")
ax[1].imshow(edges)
ax[1].set_title("Perwitt edge detection")

fig.tight_layout()
plt.show()
```
